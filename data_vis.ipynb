```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
def distribution(path, save_dir="results", dataset_type="train"):
    df = pd.read_csv(path)
    counts = [0 for _ in range(6)]
    for i in range(len(counts)):
        counts[i] = len(df[df["label"] == i])

    labels = [
        "Pants on Fire",
        "False",
        "Barely True",
        "Half True",
        "Mostly True",
        "True",
    ]

    plt.bar(labels, counts)
    plt.xlabel("Label")
    plt.ylabel("Count")
    plt.title(f"Label Distribution ({dataset_type.capitalize()})")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()

    save_path = os.path.join(save_dir, f"label_distribution_{dataset_type}.png")

    plt.show()
```

```python
distribution("data/train.csv", save_dir="results", dataset_type="train")
```

```python
distribution("data/test.csv", save_dir="results", dataset_type="test")
```

```python
distribution("data/valid.csv", save_dir="results", dataset_type="validation")
```
