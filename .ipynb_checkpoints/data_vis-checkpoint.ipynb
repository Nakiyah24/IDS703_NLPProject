```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def distribution(path):
    df = pd.read_csv(path)
    counts = [0 for _ in range(6)]
    for i in range(len(counts)):
        counts[i] = len(df[df["label"] == i])
        
    labels = ["Pants on Fire", "False", "Barely True", "Half True", "Mostly True", "True"]
    
    plt.bar(labels, counts)
    plt.xlabel('Label')
    plt.ylabel('Count')
    plt.title('Label Distribution')
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
distribution("data/train.csv")
```

```python
distribution("data/test.csv")
```

```python
distribution("data/valid.csv")
```
