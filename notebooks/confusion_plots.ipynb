```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# MAIN PLOT
confusion_matrix_main = np.array(
    [
        [165, 112, 11, 7, 1, 7],  # Actual: Pants on Fire
        [146, 398, 48, 15, 19, 34],  # Actual: False
        [38, 81, 174, 36, 17, 14],  # Actual: Barely True
        [34, 70, 38, 149, 38, 42],  # Actual: Half True
        [19, 54, 38, 30, 138, 64],  # Actual: Mostly True
        [24, 67, 11, 15, 24, 118],  # Actual: True
    ]
)

# Class labels
class_labels = [
    "Pants on Fire",
    "False",
    "Barely True",
    "Half True",
    "Mostly True",
    "True",
]

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(
    confusion_matrix_main,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=class_labels,
    yticklabels=class_labels,
)
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Model 1: Confusion Matrix")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()

# Save the plot as an image
plt.savefig("main_confusion_matrix.png")
plt.show()
```
