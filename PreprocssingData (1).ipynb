```python
# Load the data
df = pd.read_csv("LIAR2/train.csv")
```

```python
# Split subjects by ";" and filter out NaNs
df['subject'] = df['subject'].fillna("").apply(lambda x: x.split(";"))

# Use MultiLabelBinarizer to one-hot encode
mlb = MultiLabelBinarizer()
OneHotVectors = mlb.fit_transform(df['subject'].tolist())

# Convert OneHotVectors to a DataFrame for easier analysis if needed
df_OneHot = pd.DataFrame(OneHotVectors, columns=mlb.classes_)
display(df_OneHot.head(100))

# Step 3: Flatten the list and count occurrences
subjects = [subject for sublist in df['subject'] for subject in sublist]
subjectCounts = Counter(subjects)

# Convert Counter to a dictionary
subjectCountsDict = dict(subjectCounts)

print("Subjects and Counts:")
print(subjectCountsDict)
```

```python
# Handle NaN values by filling with a placeholder and splitting (in this case, no splitting is needed)
df['state_info'] = df['state_info'].fillna("NONE").apply(lambda x: [x])  # Wrap single states into a list

# Initialize MultiLabelBinarizer and fit-transform the state_info column
mlb = MultiLabelBinarizer()
state_info_one_hot = mlb.fit_transform(df['state_info'])

# Convert the one-hot encoded data to a DataFrame
df_state_info_one_hot = pd.DataFrame(state_info_one_hot, columns=mlb.classes_)
display(df_state_info_one_hot.head())

# Step 3: Flatten the list and count occurrences
all_states = [state for sublist in df['state_info'] for state in sublist]
state_counts = Counter(all_states)

# Convert Counter to a dictionary
state_counts_dict = dict(state_counts)
print("State Vocabulary and Counts:")
print(state_counts_dict)
```
